```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
import os
import math
```

```python
path  = '/content/drive/MyDrive/DMS'
points = os.path.join(path, 'Random_Points.csv')
```

```python
df = pd.read_csv(points, header=0, encoding='latin1')# 'latin1' or "cp1252"
#df.to_csv(os.path.join(path,'points_utf8.csv'), index=False, encoding='utf-8')
```

```python
df.head()
```

```python
# Use regex to extract degree, minute, second
df[['Degree', 'Minute', 'Second']] = df['Location'].str.extract(r'(\d+)°\s+(\d+)\'\s+([\d\.]+)"')
# Convert to proper types
df['Degree'] = df['Degree'].astype(int)
df['Minute'] = df['Minute'].astype(int)
df['Second'] = df['Second'].astype(float)
df.head()
```

```python
# Add Columns for Decimal Degree
df['Location_Decimal_D'] = df['Degree'] + df['Minute']/60 + df['Second']/3600
df.head()
```

```python
# df = df.drop(['Location','Degree','Minute','Second'], axis=1)
# df.head()
df.to_csv(os.path.join(path,'Location_Decimal_Degree.csv'), index=False, encoding='utf-8')
```
