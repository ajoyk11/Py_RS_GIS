```python
from shapely.geometry import Polygon
import geopandas as gpd
import os

# Define the bounding box coordinates min_lon, max_lon & min_lat, max_lat and Create the rectangle polygon
rectangle = Polygon([(77,28),(77,32),(82,32),(82,28),(77, 28)]) # the fifth one (same as the first to Close the polygon)

# Create a GeoDataFrame with EPSG:4326 (WGS 84)
gdf = gpd.GeoDataFrame(index=[0], crs="EPSG:4326", geometry=[rectangle])

# Define output directory and filename
output_dir = r"/content/Drive/MyDrive/shp"
output_file = os.path.join(output_dir, "UK_Rec_Bound_84.shp")

# Save as Shapefile
gdf.to_file(output_file, driver="ESRI Shapefile")
```
